```python
import p5_util
dict_param_sequence = p5_util.object_load('./data/dict_param_sequence.dump')
dict_param_sequence
```

```python
import p5_util
import test_datapreparator

data_type = 'train'
dict_param_sequence['data_type'] = data_type
filename = test_datapreparator.datapreparator_filename(data_type,dict_param_sequence, step=None)
dataPreparator_v2_train= p5_util.object_load(filename)

data_type = 'valid'
dict_param_sequence['data_type'] = data_type
filename = test_datapreparator.datapreparator_filename(data_type,dict_param_sequence, step=None)
dataPreparator_v2_valid= p5_util.object_load(filename)
```

```python
import numpy as np

y_train = dataPreparator_v2_train.y
len_train = len(y_train)

#---------------------------------------------------------
# Get a random sample from train dataset with same length then valid dataset
#---------------------------------------------------------
y_valid = np.array(dataPreparator_v2_train.df_data.target.sample(len_train))
#y_train = dataPreparator_v2_train.y

#---------------------------------------------------------
# Check distribution of toxic texts from random sample
#---------------------------------------------------------
print("Percentage of toxic texts in sample = {}%".format(np.where(y_train>0.)[0].shape[0]*100/len_train))

#---------------------------------------------------------
# Compute MAE issued from a random prediction
#---------------------------------------------------------
mae = (np.abs(y_valid-y_train)).mean()
print("Reference MAE  = {}".format(mae))
rmse = ((y_valid-y_train)**2).mean()
print("Reference MSE  = {}".format(rmse))
```

# Results

#### CNN Keras configuration from DataGenerator

```python
import p9_util_config

#---------------------------------------------------------------
# Update of generic paramaters depending of DataGenerator 
#---------------------------------------------------------------
dict_param_keras = p9_util_config.dict_param_keras_cnn['dict_param_keras'].copy()
dict_param_keras['input_dim']          = train_generator.get_params()['keras_input_dim']
dict_param_keras['batch_size']         = train_generator.get_params()['batch_size']
dict_param_keras
p9_util_config.dict_param_keras_cnn['dict_param_keras'] = dict_param_keras.copy()
p9_util_config.dict_param_keras_cnn
```

#### Build of keras CNN model

```python
import p9_util_keras

model = p9_util_keras.keras_cnn_build(**p9_util_config.dict_param_keras_cnn)
model.summary()
```

#### Model evaluation

```python
import p9_util_config
# Train and evaluate model on dataset
epochs = p9_util_config.dict_param_keras_cnn['dict_param_keras']['nb_epoch']
verbose = p9_util_config.dict_param_keras_cnn['dict_param_keras']['verbose']

history = model.fit_generator(generator=train_generator,
                              validation_data=valid_generator,
                              use_multiprocessing=False,
                              workers=1, 
                              verbose=verbose, 
                              epochs=epochs)
```

```python
import p9_util_keras
import p9_util_config
core_name = p9_util_keras.build_model_core_name(p9_util_config.dict_param_keras_cnn)
print("")
print(core_name)
```
